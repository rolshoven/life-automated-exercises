Welcome to the fourth challenge! The goal of this exercise is to generate a Pub Quiz Crew that can automatically generate pub quiz questions based on random wikipedia articles. Have fun!

## Setup

Before you execute this cell, make sure to provide the environment variables `OPENAI_API_BASE`, `OPENAI_MODEL_NAME`, and `OPENAI_API_KEY` in the `.env` file.

```python
import os
from dotenv import load_dotenv
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
from crewai import Agent, Crew, Task
from crewai_tools import BaseTool

load_dotenv(override=True)

assert "OPENAI_MODEL_NAME" in os.environ, "No model specified in .env file!"
print("Using the following LLM model:", os.environ.get("OPENAI_MODEL_NAME"))
```

## Tools

```python
# TODO: implement the tools for your pub quiz crew
class wikipediaQuery(BaseTool):
    name: str = "get wikipedia page content"
    description: str = "fetch content from wikipedia"

    def _run(self, searchTerm: str) -> str:
        wikipedia = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())
        return wikipedia.run(searchTerm);
    
    @staticmethod
    def cache_function(*args):
        return False
```

## Agents

```python
# TODO: implement the agents for your pub quiz crew

pubquiz_generator = Agent(
    role="Senior Content Creator which writes questions for pub quizes, based on wikipedia articles",
    goal="Write 5 questions based on wikipedia articles on the general topic of {user_topic} which can be answered by swiss pub visitors",
    backstory="A knowledgeable expert in writing pubquiz questions",
    tools=[wikipediaQuery()],
    cache=False,
    verbose=True,
)
```

## Tasks

```python
# TODO: implement the tasks for your crew
wikipedia_fetch_task = Task(
    description="Load content from wikipedia",
    expected_output="markdown version of a wikipedia article",
    agent=pubquiz_generator,
)

generate_questions = Task(
    description="Use the wikipedia article and compose a good and challaging question. The questions should have proposed answers, only one is correct. Make sure, the questions are related to switzerland",
    expected_output="Short and simple list of questions in german",
    agent=pubquiz_generator,
)
```

## Kicking off the Crew

```python
# TODO: Run the crew
crew = Crew(agents=[pubquiz_generator], tasks=[wikipedia_fetch_task, generate_questions], cache=False)

request = input("Topic: ")

result = crew.kickoff(inputs={
    "user_topic": request
})

print("reply:", result.raw)
```
